```python
from langchain.prompts import PromptTemplate
from langchain.chains.llm import LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler

def gpt(message):
    prompt = PromptTemplate(
        template="""{input}""",
        input_variables=["input"]
    )
    llm = ChatOpenAI(temperature=0, streaming=True, callbacks=[StreamingStdOutCallbackHandler()])
    chain = LLMChain(llm=llm, prompt=prompt)
    chain({"input": message})

######

gpt("""**Slice City Menu**

*Specialty Pizzas:*  
- Margherita Pizza - Fresh tomato sauce, mozzarella cheese, basil leaves. Price: 5.50€
- Veggie Supreme Pizza - Bell peppers, onions, mushrooms, olives, mozzarella cheese. Price: 6€

*Classic Pizzas:*  
- Pepperoni Pizza - Pepperoni slices, tomato sauce, mozzarella cheese. Price: 7.50€
- BBQ Chicken Pizza - Grilled chicken, BBQ sauce, red onions, mozzarella cheese. Price: 8€

*Gourmet Pizzas:*
- Mediterranean Delight Pizza - Feta cheese, spinach, sun-dried tomatoes, Kalamata olives, mozzarella cheese. Price: 9.50€
- Meat Lovers Pizza - Pepperoni, sausage, bacon, ham, tomato sauce, mozzarella cheese. Price: 10€

*Sides:*
- Garlic Breadsticks (4 pieces) with marinara sauce. Price: 3.50€
- Garden Salad - Mixed greens, cherry tomatoes, cucumbers, carrots, choice of dressing. Price: 4€

*Beverages:*
- Soft Drinks (Coke, Sprite, Fanta) - 2€
- Bottled Water - 1.50€

Note: Prices are subject to change.

What pizzas can i get for less than 7€ at Slice City?""")


```
